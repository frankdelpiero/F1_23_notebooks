```python
"""Driver Laptimes Scatterplot
==============================

Plot a driver's lap times in a race, with color coding for the compounds.
"""

import fastf1
import fastf1.plotting
import seaborn as sns
from matplotlib import pyplot as plt

# The misc_mpl_mods option enables minor grid lines which clutter the plot
fastf1.plotting.setup_mpl(misc_mpl_mods=False)

###############################################################################
# Load the race session.

race = fastf1.get_session(2023, "Qatar", 'S')
race.load()
```

```python

import numpy as np
import pandas
driver_laps = race.laps.pick_driver("PIA").pick_quicklaps().dropna(subset=['LapTime']).reset_index()

# Crear la figura y los ejes
fig, ax = plt.subplots()

ax.scatter(x=driver_laps['LapNumber'].astype(int),y=driver_laps['LapTime'],color="red")
plt.title("Piastri's laps")
plt.xlabel("Number of laps")
plt.ylabel("Lap Time")
plt.show()
```
